### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

# 存取檔案

with open('./data/64_72hr_CH.xml', encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 用迴圈存取我們的資訊
areas = doc['cwbopendata']['dataset']['locations']['location']
print(len(areas))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

for area in areas:
    print(area['locationName']+","+area['weatherElement'][0]['time'][0]['dataTime']+","+area['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

for time in areas[0]['weatherElement'][0]['time']:
    print(time['dataTime']+","+time['elementValue']['value'])
```
