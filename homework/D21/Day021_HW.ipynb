* 能夠利用 Request + BeatifulSour 撰寫爬蟲，並存放到合適的資料結構


## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
# 打開瀏覽器

from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import datetime
import time
import re

def browser_pagedown(_actions):
    time.sleep(1)
    _actions.send_keys(Keys.PAGE_DOWN)
    _actions.perform()

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

str_date_yesterday=(datetime.datetime.today()+datetime.timedelta(days=-1)).strftime("%Y/%m/%d")
bool_yesterday=False
rep=re.compile(f'{str_date_yesterday}.*')

actions = ActionChains(browser)

while (bool_yesterday==False):
    soup = BeautifulSoup(browser.page_source, "html5lib")
    items = soup.find_all('span',text=rep)
    if (len(items)>0):
        bool_yesterday=True
    else:
        browser_pagedown(actions)
        
soup = BeautifulSoup(browser.page_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_="date").text.find(str_date_yesterday)>=0):
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
str_datetime_before_2_hours=(datetime.datetime.today()+datetime.timedelta(hours=-2)).strftime("%Y/%m/%d %H:%M")
print(str_datetime_before_2_hours)
for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_="date").text<=str_datetime_before_2_hours):
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

search_date=datetime.datetime.today()+datetime.timedelta(days=-3)
year=search_date.year
month=search_date.month
day=search_date.day

str_date_search_date=search_date.strftime("%Y/%m/%d")
str_datetime_max_search=str_date_search_date+" 17:00"
str_datetime_min_search=str_date_search_date+" 15:00"

s_year = Select(browser.find_element_by_id('selY'))
s_year.select_by_value(str(year))

s_month = Select(browser.find_element_by_id('selM'))
s_month.select_by_value(str(month))

s_day = Select(browser.find_element_by_id('selD'))
s_day.select_by_value(str(day))

button=browser.find_element_by_id('button')
button.click()

bool_reach=False
rep=re.compile(f'{str_date_search_date}.*')

while (bool_reach==False):
    soup = BeautifulSoup(browser.page_source, "html5lib")
    items = soup.find_all('span',text=rep)
    if (len(items)>0):
        bool_reach=items[-1].text<str_datetime_min_search
    if (bool_reach==False):
        browser_pagedown(actions)

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_="date").text<=str_datetime_min_search):
        break
    if (d.find(class_="date").text<=str_datetime_max_search):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        
print("finished")
```
