### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
class News:
    def __init__(self, _date, _tag, _title):
        self.date=_date
        self.tag=_tag
        self.title=_title
dict={}
for d in soup.find(class_="part_list_2").find_all('h3'):
    category=d.find(class_="tag").text
    if (dict.get(category)==None):
        dict.setdefault(category,[])
    dict[category].append(News(d.find(class_="date").text,category,d.find_all('a')[-1].text))
```

### ③ 哪一個類別的文章最多

```python
max_category=''
max_category_count=0
for _category in dict:
    count=len(dict[_category])
    if (count>max_category_count):
        max_category_count=count
        max_category=_category
print(f"max category is {max_category} ({max_category_count} items)")
```
